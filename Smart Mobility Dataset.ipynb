## Smart Mobility Dataset

# 1.Introduction:
### The Smart Mobility Dataset is a detailed collection of data points focused on urban transportation and traffic conditions. It captures various aspects of mobility, environmental factors, and traffic dynamics. The dataset includes the following key attributes:

* Timestamp: The date and time of each observation.
Latitude and Longitude: Geographic coordinates indicating the location of the observation.
* Vehicle_Count: The number of vehicles recorded at a specific location.
* Traffic_Speed_kmh: The average speed of vehicles in kilometers per hour.
* Road_Occupancy_%: The percentage of road space occupied by vehicles.
* Traffic_Light_State: The state of the traffic light (e.g., Red, Yellow, Green).
* Weather_Condition: Weather conditions during the observation (e.g., Clear, Rain, Fog, Snow).
Accident_Report: Indicates whether an accident occurred (0 for no, 1 for yes).
* Sentiment_Score: A numerical score reflecting public sentiment about traffic conditions.
* Ride_Sharing_Demand: The demand for ride-sharing services at the time.
* Parking_Availability: The number of available parking spaces.
* Emission_Levels_g_km: The level of emissions in grams per kilometer.
* Energy_Consumption_L_h: Energy consumption in liters per hour.
* Traffic_Condition: A qualitative label for traffic conditions (e.g., High, Medium, Low).
This dataset is valuable for analyzing traffic patterns, understanding the impact of environmental factors on mobility, and optimizing urban transportation systems. It can be used for predictive modeling, traffic management, and smart city planning.

## inplace=True vs inplace=False in Pandas

### Understanding the `inplace` Parameter in Pandas Functions

When using pandas functions like `.dropna()`, `.fillna()`, `.drop()`, etc., the `inplace` parameter decides whether to modify the DataFrame directly or return a new one.

#### `inplace=True` (Modifies the Original DataFrame)
- Changes are made directly to the existing DataFrame.
- No need to assign the result to a new variable.

#### `inplace=False` (Creates a New DataFrame)
- The function returns a new DataFrame with changes.
- You must assign it to a variable if you want to keep the changes.

```python
df.dropna(inplace=True)
print("After dropna (inplace=True):")
df.head()
```

```python
df.dropna(inplace=False)
print("After dropna (inplace=False):")
df.head()
```

### Visualization of Total Vehicle Count by Traffic Condition

```python
import matplotlib.pyplot as plt
df = pd.read_csv("smart_mobility_dataset.csv")
traffic_summary = df.groupby("Traffic_Condition")["Vehicle_Count"].sum()
traffic_summary.plot(kind="bar", color="blue", edgecolor="black")
plt.title("Total Vehicle Count by Traffic Condition")
plt.xlabel("Traffic Condition")
plt.ylabel("Total Vehicle Count")
plt.xticks(rotation=45)
plt.show()
```

## Analyze Traffic Conditions Over Time
* This analysis examines how traffic conditions (High, Medium, Low) change over time using the dataset. By grouping data by time intervals (e.g., hourly or daily) and analyzing trends, it helps identify patterns in traffic behavior, such as peak traffic hours or conditions influenced by external factors like weather. This provides insights into traffic flow and helps in planning or decision-making.



```python
traffic_condition_counts = df['Traffic_Condition'].value_counts()
traffic_condition_counts.plot(kind="bar", color="skyblue", edgecolor="black")
plt.title("Traffic Conditions Distribution")
plt.xlabel("Traffic Condition")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

## Pie Chart Visualization of Traffic Condition Distribution
* This means creating a pie chart to show how traffic conditions (High, Medium, Low) are distributed in the dataset. Each slice of the pie represents a traffic condition, and its size shows the percentage of that condition compared to the total. This helps to easily understand which traffic condition occurs most or least frequently in the data.

```python
traffic_condition_counts = df['Traffic_Condition'].value_counts()
traffic_condition_counts.plot(kind='pie', autopct='%1.1f%%', legend=True, title='Traffic Condition Distribution')
plt.ylabel("")  # Remove y-axis label for better visualization
plt.show()
```

### Histogram of Vehicle Count Distribution
* This code generates a histogram to visualize the distribution of vehicle counts in the dataset. The histogram shows how frequently different vehicle count values occur, with a kernel density estimate (KDE) overlay to highlight the data's overall shape. The x-axis represents the vehicle count, while the y-axis shows the frequency of occurrences. This visualization helps identify patterns, such as the most common vehicle count range or any outliers in the data.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
df = pd.read_csv("smart_mobility_dataset.csv")
sns.set_style('darkgrid', {"grid.color": "0.7", "grid.linestyle": "-"})
sns.histplot(df['Vehicle_Count'], kde=True, color='purple', edgecolor='black')
plt.xlabel("Vehicle Count")
plt.ylabel("Frequency")
plt.title("Distribution of Vehicle Count")
plt.show()
```

### Correlation Heatmap of Numeric Features 
* This code creates a heatmap to show the relationships between numeric features in the dataset. It helps identify which features are strongly or weakly related, making it easier to understand patterns in the data.

```python
correlation_matrix = df.corr(numeric_only=True)
sns.heatmap(correlation_matrix.round(2), annot=True, cmap=sns.color_palette("vlag", as_cmap=True))
plt.title("Correlation Heatmap of Numeric Features")
plt.show()
```
